# Sales Forcasting System

# Monthly Sales History

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = {
    'Month_Index' : [1, 2, 3, 4, 5, 6],
    'Sales' : [18000, 25000, 21000, 38000, 23000, 45000]
}
df = pd.DataFrame(data)
```

# 2) Prepare The Model
* X = Month Index (Features)
* y = Sales (Target)

```python
X = df[['Month_Index']].values   
y = df['Sales'].values

Model = LinearRegression()
Model.fit(X, y)
```

```python
Month_Range = np.arange(1, 13).reshape(-1, 1)
Predicted_Sales_Line = Model.predict(Month_Range)

plt.figure(figsize=(8,5))
plt.scatter(X, y, color='blue', label='Actual Sales')
plt.plot(Month_Range, Predicted_Sales_Line, label='Trend Line', linewidth=2)

plt.xlabel("Month Index")
plt.ylabel("Sales")
plt.title("Sales Trend and Forecast")
plt.legend()
plt.grid(True)
plt.show()
```

# 3) Predict Next Month 
* Next Month 

```python
Next_Month = np.array([[7]])
Prediction = Model.predict(Next_Month)
print(f'Predicted Sales For The Next Month = ${Prediction[0]:.2f}')
```
